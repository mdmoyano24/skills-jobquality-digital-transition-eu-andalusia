```python
import pandas as pd
import statsmodels.formula.api as smf
from pathlib import Path

# Asegura carpetas de salida
Path("outputs").mkdir(exist_ok=True)
Path("outputs/figures").mkdir(parents=True, exist_ok=True)
Path("outputs/stata").mkdir(parents=True, exist_ok=True)  # usamos esta para dejar el .txt "auditable"

# Carga el sample y guarda panel.csv (misma estructura que usará todo)
df = pd.read_csv("data/processed/panel_sample.csv")
df.to_csv("data/processed/panel.csv", index=False)

# FE "a lo sencillo" con dummies de región y año; SE agrupados por región
mod = smf.ols("telework_share ~ temp_share + broadband + C(region) + C(year)", data=df)\
         .fit(cov_type="cluster", cov_kwds={"groups": df["region"]})

# Muestra resumen y guarda evidencia en el repo
print(mod.summary())
Path("outputs/stata/fe_results_python.txt").write_text(mod.summary().as_text(), encoding="utf-8")
```

```python
import matplotlib.pyplot as plt

d_and = df[df["region"]=="Andalucía"].sort_values("year")
plt.figure()
plt.plot(d_and["year"], d_and["telework_share"], marker="o")
plt.title("Telework share — Andalucía")
plt.xlabel("Year"); plt.ylabel("% employees")
plt.savefig("outputs/figures/telework_andalucia.png", bbox_inches="tight")
plt.show()
```
